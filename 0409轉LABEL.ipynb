```python
import numpy as np
import pandas as pd
import random
import seaborn as sns
import tensorflow
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from tensorflow.keras import models
from tensorflow.keras import layers
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
a=pd.read_csv('0409篩至378萬.csv')
df=pd.DataFrame(a)
df
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
label_feature01= labelencoder.fit_transform(df['club_member_status'])
feature01=pd.DataFrame(label_feature01)
feature01.columns=['club_member_statusL']

label_feature02= labelencoder.fit_transform(df['fashion_news_frequency'])
feature02=pd.DataFrame(label_feature02)
feature02.columns=['fashion_news_frequencyL']

label_feature03= labelencoder.fit_transform(df['product_group_name'])
feature03=pd.DataFrame(label_feature03)
feature03.columns=['product_group_nameL']

label_feature04= labelencoder.fit_transform(df['graphical_appearance_name'])
feature04=pd.DataFrame(label_feature04)
feature04.columns=['graphical_appearance_nameL']

label_feature05= labelencoder.fit_transform(df['colour_group_name'])
feature05=pd.DataFrame(label_feature05)
feature05.columns=['colour_group_nameL']

df['club_member_status']=feature01['club_member_statusL']
df['fashion_news_frequency']=feature02['fashion_news_frequencyL']
df['product_group_name']=feature03['product_group_nameL']
df['graphical_appearance_name']=feature04['graphical_appearance_nameL']
df['colour_group_name']=feature05['colour_group_nameL']

df
```
